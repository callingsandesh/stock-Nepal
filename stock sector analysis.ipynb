```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
data=pd.read_csv("datasets/total_listed_stock.csv")
```

```python
pd.set_option('display.max_rows', None)
```

```python
data
```

```python

sns.catplot(x='Sector',kind='count',data=data, order=data['Sector'].value_counts().index)
plt.xticks(rotation=90)
plt.savefig('sector_counts_decending.pdf')
plt.show()
```
